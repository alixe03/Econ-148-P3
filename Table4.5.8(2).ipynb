```python
import pandas as pd
import numpy as np
from linearmodels import PanelOLS
from stargazer.stargazer import Stargazer
```

```python
muni = pd.read_stata('mexico_muni_compressed2.dta')
muni
```

```python
state = pd.read_stata('mexico_state_jebo.dta')
state
```

```python
#We create a variable date to capture both year and month effects, using the first day of the month for each date
muni['day'] = 1
muni['Date'] = pd.to_datetime(dict(year=muni.year, month=muni.month, day=muni.day))
muni
```

```python
#TABLE 4 COLUMN 5
#We create different tables each time because the dataset is very big and creates problems with storing memory
table_4_hom_df = muni[muni['year'] <= 2006].dropna(subset = ['hom_rate', 'temp_pop', 'temp_income'])
table_4_hom_df = table_4_hom_df.set_index(['muni_code', 'Date'])
table_4_hom_df = table_4_hom_df[['state', 'hom_rate', 'popw', 'temp_pop', 'precip_pop', 'temp_income']]
table_4_hom_df
```

```python
mod_table_4 = PanelOLS.from_formula('hom_rate ~ temp_pop + precip_pop + temp_income + TimeEffects + EntityEffects', data=table_4_hom_df, weights= table_4_hom_df['popw'])

result_hom = mod_table_4.fit(cov_type='clustered', clusters = table_4_hom_df['state'])

print(result_hom.summary)
```

```python
#TABLE 4 COLUMN 6
table_4_hom_df = muni[muni['year'] <= 2006].dropna(subset = ['hom_rate', 'temp_pop', 'temp_gini'])
table_4_hom_df = table_4_hom_df.set_index(['muni_code', 'Date'])
table_4_hom_df = table_4_hom_df[['state', 'hom_rate', 'popw', 'temp_pop', 'precip_pop', 'temp_gini']]
table_4_hom_df
```

```python
mod_table_4 = PanelOLS.from_formula('hom_rate ~ temp_pop + precip_pop + temp_gini + TimeEffects + EntityEffects', data=table_4_hom_df, weights= table_4_hom_df['popw'])

result_hom1 = mod_table_4.fit(cov_type='clustered', clusters = table_4_hom_df['state'])

print(result_hom1.summary)
```

```python
#TABLE 4 COLUMN 8
table_4_hom_df = muni[muni['year'] <= 2006].dropna(subset = ['hom_rate', 'temp_pop', 'temp_avg_temp'])
table_4_hom_df = table_4_hom_df.set_index(['muni_code', 'Date'])
table_4_hom_df = table_4_hom_df[['state', 'hom_rate', 'popw', 'temp_pop', 'precip_pop', 'temp_avg_temp']]
table_4_hom_df
```
